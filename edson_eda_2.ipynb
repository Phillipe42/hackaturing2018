### Verificar se o sexo, idade ou preÃ§o, influencia o valor do desconto

```python
kdims=['cnpj', 'prestador', 'uf', 'id_beneficiario', 'sexo', 'data_nascimento', 'id_conta', 'cid', 
       'crm_solicitante', 'cbos_solicitante', 'cbos_executante', 'data_entrada', 'data_saida', 'data_item', 
       'senha', 'tipo_guia', 'tipo_item', 'carater_atendimento', 'servico', 'descricao_despesa', 
       'quantidade', 'valor_item', 'valor_cobrado', 'valor_pago', 'ano_mes', 'age']
ds = hv.Dataset(df, vdims=['discount'], kdims=kdims)
```

###### por sexo

```python
%%opts Distribution [height=400 width=500]
%%opts Overlay [show_legend=True legend_position='right']
hv.Distribution(df[df['sexo']=='M']['discount'], label='Male') * \
hv.Distribution(df[df['sexo']=='F']['discount'], label='Female')
```

###### Por valor do item

```python
%%opts Scatter [height=400 width=500] (size=4)
hv.Scatter(df[['valor_item', 'discount']])
```

```python
%%opts Distribution [height=400 width=400]
dists = {}
for valor in np.linspace(0,1000,20):
    dists[valor] = hv.Distribution(df[df['valor_item']>=valor]['discount']).options(bandwidth=0.25).\
                        redim(discount={'range':(0,1)})
hv.HoloMap(dists)
```

###### Por idade

```python
%%opts Scatter [height=400 width=400 tools=['hover']] (size=4)
hv.Scatter(df[['age', 'discount']])
```

```python
%%opts Distribution [height=400 width=500]
dists = {}
for idade in np.linspace(0,80,20):
    dists[idade] = hv.Distribution(df[(df['age']>=idade)&(df['age']<=idade+10)]['discount']).options(bandwidth=0.25).\
                        redim(discount={'range':(0,1)})
hv.HoloMap(dists)
```

###### Verificar pela data do procedimento

```python
%%opts Distribution [height=400 width=500]
dists = {}
for data in np.sort(df['ano_mes'].unique())[1:-1]:
    dists[data] = hv.Distribution(df[(df['ano_mes']==data)]['discount']).options(bandwidth=0.5).\
                        redim(discount={'range':(0,1)}, Density={'range':(0,5)})
hv.HoloMap(dists)
```
